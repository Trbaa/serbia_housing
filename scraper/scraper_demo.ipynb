```python
import time
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
```

```python
BASE_URL = "https://www.halooglasi.com/"
START_URL = "https://www.halooglasi.com/nekretnine/prodaja-stanova/beograd"
HEADERS = {
    "User-Agent": "Mozilla/5.0"
}
```

```python
response = requests.get(START_URL)
print(response)
```

```python
soup = BeautifulSoup(response.text,'html.parser')
```

```python
card_title = soup.find_all('h3')
for link in card_title:
    card_title_link = link.find('a')
    print(card_title_link['href'])
```

```python
##OVAKO DOBIJAM CENU NEKRETNINE
```

```python
cards = soup.find_all('div', class_='product-item')

for card in cards:
    price_tag = card.find('span', {'data-value': True})
    
    if price_tag:
        price = price_tag['data-value']
        print(price)
```

```python
#BROJ SVIH STRANICA SA SAJTA
```

```python
def get_pages(pages_num):
    last_page = pages_num
    pages_list = []
    page = 1
    
    while page <= last_page:
        url = f"{START_URL}?page={page}"
        pages_list.append(url)
        page +=1
        
    return pages_list
```

```python
pages_list = get_pages(2)

for page in pages_list:
    response = requests.get(page)
    soup = BeautifulSoup(response.text,'html.parser')
    
    card_title = soup.find_all('h3')
    
    for link in card_title:
        card_title_link = link.find('a')
        print(card_title_link['href'])
```

```python
card_title = soup.find_all('h3')
for link in card_title:
    card_title_link = link.find('a')
    new_url = BASE_URL + card_title_link['href']  
    response = requests.get(new_url, headers=HEADERS)
    soup_detail = BeautifulSoup(response.text, 'html.parser')

    title = soup_detail.find('span', {'data-code': 'Title'})
    price = soup_detail.find(class_='offer-price-value')
    print(title)
    print(price)
```
